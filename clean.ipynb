```python
import pandas as pd

# 1. CSV file ka path (original file)
file_path = 'F:\\InternShip\\Booking.com-Data-Scraping\\pakistan_home_all_cities.csv'  # Apna file ka exact path likho

# 2. CSV load karo
df = pd.read_csv(file_path)

# 3. "Scored " text remove karke numeric me convert karo
df['Rating'] = df['Rating'].str.replace('Scored ', '', regex=False).astype(float)

# 4. Price column ko remove karo (agar exist karta hai)
if 'Price' in df.columns:
    df.drop(columns=['Price'], inplace=True)

# 5. Same file overwrite karke save karo
df.to_csv(file_path, index=False)

print(f"File '{file_path}' updated successfully! 'Rating' cleaned & 'Price' column removed.")
```

```python
import pandas as pd

# 1. Apni 3 CSV file paths likho
file1 = 'F:\\InternShip\\Booking.com-Data-Scraping\\pakistan_apartment_all_cities.csv'
file2 = 'F:\\InternShip\\Booking.com-Data-Scraping\\pakistan_home_all_cities.csv'
file3 = 'F:\\InternShip\\Booking.com-Data-Scraping\\pakistan_hotel_all_cities.csv'

# 2. CSV files ko read karo
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
df3 = pd.read_csv(file3)

# 3. Combine (vertically append) karo
combined_df = pd.concat([df1, df2, df3], ignore_index=True)

# 4. Nayi file ka naam set karo
output_file = 'property_booking.csv'

# 5. Save karo
combined_df.to_csv(output_file, index=False)

print(f"Combined CSV saved as {output_file}")
```
